```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from math import pi

%matplotlib inline
sns.set(style="whitegrid")
```

```python
# 1️⃣ Veri yükleme
df = pd.read_csv('FC26_20250921.csv', low_memory=False)
print(df.head())
print(df.info())
print(df.describe())
print(df.columns.tolist())
```

```python
# 2️⃣ Temel temizleme
# Önemli sayısal sütunlar
num_cols = ['overall','potential','age','height_cm','weight_kg','value_eur','wage_eur',
            'pace','shooting','passing','dribbling','defending','physic']  

# Eksik değerleri çıkar
df = df.dropna(subset=num_cols)
```

```python
# 3️⃣ Genel istatistikler
print(df[num_cols].describe())
```

```python
# 4️⃣ Top 10 Oyuncu (Overall)
top_players = df.nlargest(10, 'overall')[['short_name','club_name','nationality_name','overall','potential']]
plt.figure(figsize=(10,6))
sns.barplot(x='overall', y='short_name', data=top_players, palette='viridis')
plt.title('FC26 - En Yüksek Overall Oyuncular')
plt.xlabel('Overall')
plt.ylabel('Oyuncu')
plt.show()
```

```python
# 5️⃣ Pozisyon bazlı ortalama overall
plt.figure(figsize=(12,6))
df['primary_position'] = df['player_positions'].str.split(',').str[0]  
pos_overall = df.groupby('primary_position')['overall'].mean().sort_values(ascending=False)
sns.barplot(x=pos_overall.values, y=pos_overall.index, palette='coolwarm')
plt.title('Pozisyon Bazında Ortalama Overall')
plt.xlabel('Ortalama Overall')
plt.ylabel('Pozisyon')
plt.show()
```

```python
# 6️⃣ Korelasyon matrisi
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Sayısal Özellikler Arası Korelasyon')
plt.show()
```

```python
# 7️⃣ Radar Chart - Oyuncu Profili
skill_cols = ['pace','shooting','passing','dribbling','defending','physic']

def player_radar(player_name):
    player = df[df['short_name'] == player_name].iloc[0]
    values = [player[col] for col in skill_cols]
    values += values[:1]  # döngü

    angles = [n / float(len(skill_cols)) * 2 * pi for n in range(len(skill_cols))]
    angles += angles[:1]

    plt.figure(figsize=(6,6))
    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], skill_cols, color='grey', size=8)
    ax.plot(angles, values, linewidth=2, linestyle='solid')
    ax.fill(angles, values, 'skyblue', alpha=0.4)
    plt.title(f"{player_name} Profil")
    plt.show()

player_radar('J. Bellingham')
```

```python
# 8️⃣ Kümelenme Analizi (oyuncu tipleri)
X = df[skill_cols].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(10,6))
sns.scatterplot(x='pace', y='dribbling', hue='cluster', data=df.sample(5000), palette='Set2', alpha=0.7)
plt.title('Pace vs Dribbling ile Kümelenmiş Oyuncular (Örnek 5k)')
plt.xlabel('Pace')
plt.ylabel('Dribbling')
plt.show()
```

```python
# 9️⃣ Benzer Oyuncu Önerisi
def recommend_similar(player_name, n=5):
    # Tek oyuncu satırı alınır
    player_vec = scaler.transform(df[df['short_name'] == player_name][skill_cols].iloc[0:1])
    
    # Öklid mesafesi
    df['dist'] = ((X_scaled - player_vec)**2).sum(axis=1)
    
    # En yakın n oyuncu
    recommendations = df.sort_values('dist').iloc[1:n+1]  
    return recommendations[['short_name','club_name','overall','primary_position']]

print(recommend_similar('J. Bellingham', 5))
```

```python
# 10️⃣ Yaş vs Overall vs Potential Analizi
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='overall', hue='potential', data=df.sample(5000), palette='coolwarm', alpha=0.6)
plt.title('Age vs Overall (Renk: Potential)')
plt.xlabel('Yaş')
plt.ylabel('Overall')
plt.show()
```

```python
# 11️⃣ Kulüp bazında ortalama overall
club_overall = df.groupby('club_name')['overall'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=club_overall.values, y=club_overall.index, palette='magma')
plt.title('Top 10 Kulüp - Ortalama Overall')
plt.xlabel('Ortalama Overall')
plt.ylabel('Kulüp')
plt.show()
```

```python
# 12️⃣ Pozisyon bazlı skill radar chart (otomatik)
def position_radar(pos_name):
    pos_df = df[df['primary_position'] == pos_name]
    avg_skills = pos_df[skill_cols].mean().values.tolist()
    avg_skills += avg_skills[:1]
    angles = [n / float(len(skill_cols)) * 2 * pi for n in range(len(skill_cols))]
    angles += angles[:1]

    plt.figure(figsize=(6,6))
    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], skill_cols, color='grey', size=8)
    ax.plot(angles, avg_skills, linewidth=2, linestyle='solid', color='red')
    ax.fill(angles, avg_skills, 'pink', alpha=0.3)
    plt.title(f"{pos_name} Pozisyon Ortalaması")
    plt.show()

# Örnek: CAM pozisyonu
position_radar('CAM')
```

```python
# 13️⃣ Pozisyon bazlı benzer oyuncu önerisi
def recommend_similar_pos(player_name, n=5):
    player = df[df['short_name'] == player_name].iloc[0]
    pos_players = df[df['primary_position'] == player['primary_position']].copy()
    
    # StandardScaler ile normalize
    pos_scaled = scaler.transform(pos_players[skill_cols])
    player_vec = scaler.transform(pos_players[skill_cols].iloc[[pos_players.index.get_loc(player.name)]])
    
    pos_players['dist'] = ((pos_scaled - player_vec)**2).sum(axis=1)
    
    recommendations = pos_players.sort_values('dist').iloc[1:n+1]
    return recommendations[['short_name','club_name','overall','primary_position']]

# Örnek: Pozisyon bazlı öneri
print(recommend_similar_pos('J. Bellingham', 5))
```

```python
# 14️⃣ Cluster Profiling - Küme ortalama radar chart
def cluster_radar(cluster_num):
    cluster_df = df[df['cluster'] == cluster_num]
    avg_skills = cluster_df[skill_cols].mean().values.tolist()
    avg_skills += avg_skills[:1]
    angles = [n / float(len(skill_cols)) * 2 * pi for n in range(len(skill_cols))]
    angles += angles[:1]

    plt.figure(figsize=(6,6))
    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], skill_cols, color='grey', size=8)
    ax.plot(angles, avg_skills, linewidth=2, linestyle='solid', color='green')
    ax.fill(angles, avg_skills, 'lightgreen', alpha=0.3)
    plt.title(f"Cluster {cluster_num} Ortalaması")
    plt.show()

cluster_radar(0)
```
